```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset into pandas dataframe
results = pd.read_excel('C:\Austine\Blog\ML\Reference\Maths_score.xlsx')
```

```python
#Always understand the axis: labels and scale

results.plot(kind='scatter', x='No. of hours studied', y='Marks received');
plt.xlabel('No. of hours studied daily')
plt.title('Maths Score and hours spent in studying daily')
plt.show()

#We can see that the students who are spending more time studies daily have received higher score in maths
```
